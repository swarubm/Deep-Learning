# 5. Design and implement a deep learning network for classification of textual documents

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import imdb
import numpy as np
# Load and preprocess IMDb dataset
vocab_size = 10000
max_length = 200
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=vocab_size)
x_train = tf.keras.preprocessing.sequence.pad_sequences(x_train, maxlen=max_length)
x_test = tf.keras.preprocessing.sequence.pad_sequences(x_test, maxlen=max_length)
# Define the model
model = models.Sequential([
layers.Input(shape=(max_length,)),
layers.Embedding(vocab_size, 128),
layers.LSTM(64, return_sequences=False),
layers.Dense(32, activation='relu'),
layers.Dense(1, activation='sigmoid')
])
# Compile model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
# Train model
model.fit(x_train, y_train, epochs=3, batch_size=128, validation_split=0.2, verbose=1)
# Evaluate model
test_loss, test_accuracy = model.evaluate(x_test, y_test, verbose=0)
print(f"Test Accuracy: {test_accuracy:.4f}, Test Loss: {test_loss:.4f}")
# Predict on test set
y_pred = model.predict(x_test, verbose=0)
y_pred_classes = (y_pred > 0.5).astype(int).flatten()
print(f"Sample Predictions: {y_pred_classes[:5]} (True: {y_test[:5]})")
```
